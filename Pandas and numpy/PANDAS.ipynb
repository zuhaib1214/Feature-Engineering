```python
#find the most destructive death over batsman ins history of IPL
#Strike Rate=(Number of runs/Number of balls)/100
#Min batsman 200 balls in iver 16-20
```

```python
mas=de['over']>15
death=de[mas]
```

```python
death_del=death.groupby('batsman')['batsman_runs'].count()
g=death_del>200
batsman_list=death_del[g].index.tolist()
```

```python
#Run scored by all these 43 betsmen
#Balls played by these 43 betsmen
```

```python
final=de[de['batsman'].isin(batsman_list)]
```

```python
final
```

```python
final_runs=final.groupby('batsman')['batsman_runs'].sum() 
final_runs
```

```python
final_balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
final_balls
```

```python
strike_rate=(final_runs/final_balls)*100
```

```python
strike_rate
```
