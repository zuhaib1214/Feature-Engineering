# Converter

```python
df11=pd.read_csv('matches - matches.csv')
df11
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
df11=pd.read_csv('matches - matches.csv',converters={'team1':rename})
df11.head(100)
#soo first we create a function of changing the name of royal challengers bangalore into RCB
#and in read_csv() method we use converter parameter in which we pass the column whose value we
#want to change and pass the function
```
