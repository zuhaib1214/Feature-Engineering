# train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('purchased',axis=1),df['purchased'],test_size=0.2)
```

# Ordinal encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)

```

```python
X_train_ord=oe.transform(X_train)
X_test_ord=oe.transform(X_test)
```

```python
X_train_ord = pd.DataFrame(X_train_ord).rename(columns={0:'review',1:'educatioin'})
X_test_ord = pd.DataFrame(X_test_ord).rename(columns={0:'review',1:'educatioin'})
```
