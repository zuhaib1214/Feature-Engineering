```python
import pandas as pd
import seaborn as sns
```

```python
tip=sns.load_dataset('tips')
```

```python
tip.head()
```

```python
titanic=pd.read_csv('titanic.csv')
titanic.head()
```

```python
flight=sns.load_dataset('flights')
flight.head()
```

```python
iri=sns.load_dataset('iris')
iri.head()
```

# Bivariate
## Scatterplot(Numerical-Numerical )

```python
sns.scatterplot(tip['total_bill'],tip['tip'])
```

# Multivariate
## scatterplot(Numeical-Numerical)

```python
sns.scatterplot(tip['total_bill'],tip['tip'],hue=tip['sex'],style=tip['smoker'],size=tip['size'])
```

# Bivariate
## Bar Plot(Numerical-categorical)

```python
sns.barplot(titanic['Pclass'],titanic['Age'])
```

```python
sns.barplot(titanic['Pclass'],titanic['Fare'])
```

# Multivariate
## Bar Plot(Numerical-categorical)

```python
sns.barplot(titanic['Pclass'],titanic['Age'],hue=titanic['Sex'])
```

```python
sns.barplot(titanic['Pclass'],titanic['Fare'],hue=titanic['Sex'])
```

# Bivariate
## Box Plot(Numerical-categorical)

```python
sns.boxplot(titanic['Sex'],titanic['Age'])
```

```python
sns.boxplot(titanic['Sex'],titanic['Fare'])
```

# MultiVariate
## Box Plot(Numerical-categorical)

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

```python
sns.boxplot(titanic['Sex'],titanic['Fare'],hue=titanic['Survived'])
```

# Bivariate
## Dist Plot(Numerical-categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# Bivariate
## HeatMap(Categorical-categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

# Bivariate
## ClusterMaP(Categorical-categorical)

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
