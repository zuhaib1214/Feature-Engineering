# Working with dates

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1.Extract Year

```python
date['date_year']=date['date'].dt.year
date.head(10)
```

## 2.Extract Month

```python
date['date_month']=date['date'].dt.month
date.head(10)
```

## 3.Extract month name

```python
date['date_month_name']=date['date'].dt.month_name()
date.head(10)
```

## 4. Extract day

```python
date['date_day']=date['date'].dt.day
date.head(10)
```

## 5.Extract day of week

```python
date['date_dow']=date['date'].dt.dayofweek
date.head(10)
```

## 6. Extract days of week name

```python
date['date_dow_name']=date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders'],inplace=True)
date.head(10)
```

## 7. checking is day is weekend

```python
date['date_is_weekend']=np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']),1,0)
date.head(10)
```

## 8. extract week of the year

```python
date['date_week']=date['date'].dt.week
date.head(10)
```

## 9.extract quarter

```python
date['Quarter']=date['date'].dt.quarter
date.head(10)
```

## 10.extract semester

```python
date['semester']=np.where(date['Quarter'].isin([1,2]),1,2)
date.head()
```

## 11. extract time elapsed between dates

```python
import datetime
today=datetime.datetime.today()
today
```

```python
time_elapsed=today-date['date']
time_elapsed
```

```python
# if you want only days
time_elapsed.dt.days
```

```python
#if you want month passed
np.round(time_elapsed/np.timedelta64(1,'M'),0)
```
